```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Datasets/diabetes.csv')
```

```python
df
```

```python
sorted(df['pregnancies'].unique())
```

```python
logreg = LogisticRegression()
```

```python
x = np.array(df['pregnancies']).reshape(-1,1)
y = df['diabetes']
```

```python
logreg.fit(x,y)
```

```python
pregnancies = np.array(list(range(0,18))).reshape(-1,1)
pred_on_pregnancies = logreg.predict(pregnancies)
```

```python
data = {
    'pregnancies' : np.array(list(range(0,18))),
    'pred_on_pregnancies' : pred_on_pregnancies
}
```

```python
data
```

```python
pred_df = pd.DataFrame(data)
```

```python
plt.scatter(pred_df['pregnancies'], pred_df['pred_on_pregnancies'])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, mean_squared_error, mean_absolute_error
```

```python
df
```

```python
x = df[['pregnancies', 'glucose', 'diastolic', 'triceps', 'insulin', 'bmi', 'dpf', 'age']]
y = df['diabetes']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
logreg = LogisticRegression(max_iter = 200)
```

```python
logreg.fit(x_train, y_train)
```

```python
y_pred = logreg.predict(x_test)
```

```python
accuracy_score(y_test, y_pred)
```

```python
mean_squared_error(y_test, y_pred)
```

```python
mean_absolute_error(y_test, y_pred)
```
