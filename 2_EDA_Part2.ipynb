```python
import pandas as pd
```

```python
data = pd.read_csv('Datasets/homelessness.csv')
```

```python
data.head()
```

```python
# data = data.drop('Unnamed: 0', axis=1)
data.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
data['state_pop_million'] = data['state_pop']/1000000
```

```python
data.to_csv('homelessness_updated.csv')
```

```python
#conditional filters
```

```python
data['region'] == 'Pacific'
```

```python
data[data['region'] == 'Pacific']
```

```python
data[(data['state_pop']>1000000) & (data['region'] == 'Pacific')]
```

```python
data[(data['state_pop']>1000000) & ((data['region'] == 'Pacific') | (data['region'] == 'Mountain'))]
```

```python
data[(data['state_pop']>1000000) & ((data['region'] == 'Pacific') | (data['region'] == 'Mountain'))].sort_values('region')
```

```python
data[(data['state_pop']>1000000) & ((data['region'] == 'Pacific') | (data['region'] == 'Mountain'))].sort_values(['region','individuals'])
```

```python
# Where + GroupBy
data[(data['region'] == 'Pacific') | (data['region'] == 'Mountain')].groupby('region')['state_pop'].sum()
```

```python
# Where + Group By + Having
data[(data['region'] == 'Pacific') | (data['region'] == 'Mountain')].groupby('region')['state_pop'].sum()['Mountain']
```
