```python
# Left Join

# select *
# from A
# left join B
# on A.column = B.column

# Left Table = A
# Right Table = B

#Right Join 
# select *
# from A
# right join B
# on A.column = B.column

# Right Table = A
# Left Table = B


#Self Join 
# select *
# from A
# inner join A
# on A.column = B.column
```

```python
movies = pd.read_pickle('Datasets/movies.p')
```

```python
pd.merge(movies, genre, how="left", left_on="id", right_on="movie_id")
```

```python
genre.drop_duplicates()
```

```python
genre.drop_duplicates(subset=['movie_id'])
```

```python
pd.merge(movies, genre.drop_duplicates(subset=['movie_id']), how="left", left_on="id", right_on="movie_id")
```

```python
# In case of left join the ideal number of output rows should be equal to the number of rows in left table given..
# given that right table has no duplicates in the join column..
# if there are duplicates, the number of rows in output table will be more.
```
